```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

**Do the following tasks:**
* set style to '`white`'
* load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic.head()
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
tips.head()
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.**

```python
palette_names = sns.palettes.SEABORN_PALETTES.keys()
print(list(palette_names))
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')
plt.figure(figsize=(4, 3))
sns.boxplot(x="day", y="tip", data=tips, hue = 'sex', palette='deep6', fliersize=5)
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',height=4.5,ratio=5)
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

sns.displot(titanic['fare'],kde=False,bins=15,color='green', height=3.5)
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.boxplot(x='class',y='age',data=titanic,palette='pastel',hue = 'alive')
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(x='fare',y='age',ratio=5,data=titanic,height=5)
```

**Recreate the plot below. Use titanic dataset. **

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.boxplot(x='class',y='age',data=titanic,palette="Set2",legend=False, hue='class')
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4, 3))
sns.countplot(hue='class',x='sex',data=titanic, palette='deep6')
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots**

```python
# Code here and leave the next cell empty, so that you don't lose the plot
# Select only numeric columns for correlation
titanic_numeric = titanic.select_dtypes(include=['number'])

# Compute the correlation matrix
corr = titanic_numeric.corr()

# Create a heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
