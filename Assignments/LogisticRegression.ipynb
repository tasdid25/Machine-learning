```python
import pandas as pd
from matplotlib import pyplot as plt
```

**load the csv**

```python
#code here
df=pd.read_csv('titanic_full.csv')
df.head()
```

**Check the null value**

```python
#code here
df.isnull().sum()
```

**Plot the data**

```python
#code here
plt.figure(figsize=(4,3))
plt.scatter(df.Fare,df.Survived,marker='+',color='red')
```

**Separate the Features a target in x and y**

```python
#code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['Fare']], df.Survived, train_size=0.8, random_state=42)
```

**Split the train and test data**

```python
#code here
print(X_train.head(),y_train.head(),X_test.head(),y_test.head())
```

**Make the model Binary logistic regression model**

```python
#code here
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train, y_train)
```

**Check the model score**

```python
#code here
model.score(X_test,y_test)
```
