**14. We need to send new cards to the customers well before the expire, how many cards are expiring in 2019?**<br>
*Use `sum()` and `count()` and see the difference in their use :)*

```python
# Code here please
# cc_exp is given in 04/2019 format so we used str.contains which returns bool value
sum_expiring_2019 = (df['cc_exp'].str.contains('2019')).sum()

print(f"Number of credit cards expiring in 2019 (using sum()): {sum_expiring_2019}")

# Filter the DataFrame for credit cards expiring in 2019
expiring_2019 = df[df['cc_exp'].str.contains('2019')]
# Count the number of such credit cards using count()
count_expiring_2019 = expiring_2019['cc_no'].count()

print(f"Number of credit cards expiring in 2019 (using count()): {count_expiring_2019}")
```
