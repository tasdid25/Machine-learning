<h2>Categorical Variables and One Hot Encoding</h2>

```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices.csv")
df
```

Alternative Method:

```python
df_encoded = pd.get_dummies(df, columns=['town']) #Column names in the DataFrame to be encoded

df_encoded
```

<h2 style='color:purple'>Using pandas to create dummy variables</h2>

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town'], axis='columns')
#final = final[['monroe township', 'robinsville', 'west windsor', 'area', 'price']]
final
```
