### Feature Importance Method

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cardio_train.csv')
df
```

```python
x = df.drop('cardio', axis=1)
x
```

```python
x = x.drop('id', axis=1)
x
```

```python
y = df.cardio
y
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x, y)
```

```python
model.feature_importances_
```

```python
importance_col = pd.DataFrame(data = model.feature_importances_, columns = ['importance'])
importance_col
```

```python
Name_col = pd.DataFrame(data = x.columns, columns = ['Features'])
Name_col
```

```python
Features_importance = pd.concat([Name_col, importance_col], axis=1)
Features_importance
```

```python
Features_importance.nlargest(8, 'importance')
```
